```python
import gc
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

gc.collect()
ticker = ["DOGE" , "USDT"]
logs = pd.read_csv("./log/" + ticker[0] + "_log.csv", header=0,
                 names=["DTTM", "MONEY", "AMOUNT", "MONEY_TOTAL", "RATE_TOMORROW", "CLOSE_VALUE"])
logs = logs.iloc[0:]

print("%s ~ %s"%(logs["DTTM"].iloc[0].split(".")[0], logs["DTTM"].iloc[-1].split(".")[0]))
print("%.2f %s, %.0f %s"%(logs["MONEY"].iloc[-1], ticker[1], logs['AMOUNT'].iloc[-1], ticker[0]))
print("%.2f %%"%(logs['MONEY_TOTAL'].iloc[-1]/logs['MONEY_TOTAL'].iloc[0]*100-100), end=", ")
print("W %.2f"%(logs['MONEY_TOTAL'].iloc[-1] * 1100))
print("%.4f %s/%s"%(logs['CLOSE_VALUE'].iloc[-1], ticker[1], ticker[0]))

plt.clf()
plt.plot(logs['CLOSE_VALUE']/logs['CLOSE_VALUE'].iloc[0])
plt.plot(logs['MONEY_TOTAL']/logs['MONEY_TOTAL'].iloc[0])
plt.show()
```

```python
thresholds = np.zeros((len(logs['RATE_TOMORROW']),2))
thresholds[:,0] = 0.0
thresholds[:,1] = -0.0

print(logs['RATE_TOMORROW'].iloc[-1])
plt.clf()
plt.plot(np.clip(logs['RATE_TOMORROW'].to_numpy(), -1.0, 1.0))
plt.plot(thresholds)
plt.show()
```
